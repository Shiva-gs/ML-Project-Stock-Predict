```python
import pandas as pd
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
```

```python
tables = pd.read_html(url)
tables
```

```python
df_500 = tables[0]
df_500
```

```python
df_500.groupby(['GICS Sector'])['Symbol'].count()
```

```python
realEstate_list = df_500[df_500['GICS Sector']=='Real Estate']['Symbol'].to_list()
communication_list = df_500[df_500['GICS Sector']=='Communication Services']['Symbol'].to_list()
consumerStaples_list = df_500[df_500['GICS Sector']=='Consumer Staples']['Symbol'].to_list()
consumer_list =  df_500[df_500['GICS Sector']=='Consumer Discretionary']['Symbol'].to_list()
energy_list = df_500[df_500['GICS Sector']=='Energy']['Symbol'].to_list()
financials_list = df_500[df_500['GICS Sector']=='Financials']['Symbol'].to_list()
health_list = df_500[df_500['GICS Sector']=='Health Care']['Symbol'].to_list()
industrials_list = df_500[df_500['GICS Sector']=='Industirals']['Symbol'].to_list()
information_list = df_500[df_500['GICS Sector']=='Information Technology']['Symbol'].to_list()
materials_list = df_500[df_500['GICS Sector']=='Materials']['Symbol'].to_list()
utilities_list = df_500[df_500['GICS Sector']=='Utilities']['Symbol'].to_list()
```

```python
print(realEstate_list)
print(communication_list)
print(consumer_list)
print(consumerStaples_list)
print(energy_list)
print(financials_list)
print(health_list)
print(industrials_list)
print(information_list)
print(materials_list)
print(utilities_list)
```
